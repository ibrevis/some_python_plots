```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0,1,20)
y1 = x**2 + 0.01
y2 = np.cos(x)
```

```python
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "serif",
    "font.serif": ["Computer Modern Roman"], # LaTex font
})
```

```python
FontSize = 18
```

```python
plt.figure(figsize=(8,6))
# first plot
plt.plot(x,y1,color='tab:orange',
         linestyle='--',
         marker='o',
         label='first label',
         linewidth=2.0)
# second plot
plt.plot(x,y2,color='tab:green',
         linestyle='--',
         marker=None,
         label='second label',
         linewidth=2.0)
# axis labels and sizes
plt.ylabel(r'$y$ axis label',fontsize=FontSize) # "r" means raw sequence
plt.yticks(fontsize=FontSize-2)
plt.xlabel(r'$x$ axis label',fontsize=FontSize)
plt.xticks(fontsize=FontSize-2)
# axis limits
plt.xlim(0,None)
plt.ylim(0,None)
# Main title and legend
plt.title(r'Main title with a $\theta$ character',fontsize=FontSize) 
plt.legend(fontsize=FontSize, loc='lower right')
#
plt.show()
```
